## 4) OLS vs. IV

The problem with her statement is first that she only looks at the factors uncorrelated with price. But, in an OLS model, simply only including the factors assumed to be uncorrelated with a certain demand shock does not automatically imply that the model is unbiased. We fear that we include regressors which will in turn be influenced by the model itself. If we can guarantee that no reverse causality mechanism operates, we may think that the variation is indeed exogeneous. However, there might be other correlated unobservables which have an influence on both price and quantity. Secondly, the argument that better forecasting models should always be preferred to less precise ones is also not true, potentially due to the oversatisfied variables problem, which states that, when using inappropriate measures of prediction accuracy, then addition of additional variables will automatically increase accuracy of the model, regardless of the logical relationship between regressor and dependend variable. Further, if you only include approximations of prices based on factors uncorrelated to demand shocks, then you only have an approximated result based on price. You might not be able to tell the effect of price on demand based on sole price increases as you know that you only included factors related to price, but not price itself. 

## 5) Extended Analysis of demand equation

***Situation:**

- $q_i^D(p) = a_0 + a_1p_i + u_i$ 
- Monopolist with MC = 0 (a) and MC = $c_i$ (b)
- $E(u_i) = 0$ and $E(c_i | u_i) = 0$

### a) Describe the behavior of the OLS estimate of quantiy on price: 

We need to show that the OLS estimator, given by $a_1$ is biased. In order to show this, we first need to show that: 

$$
\hat a_1 = \frac{Cov(q_i^*, p_i^*)}{var(p_i^*)}
$$

Which is the bivariate OLS estimator in our case. 

To do so, we follow several steps: 

1) Calculate the optimal price and quantity for the monopolist

The optimal profit for the monopolist is given by $(p_i - c_i)*q_i^D$ where $c_i = 0$ in the first case. As a consequence, our FOC becomes: 

$$
max \underbrace{\pi}_{p_i} = a_0 + 2a_1p_i + u_i = 0
$$

And therefore, our optimal price for the monopolist is: 

$$
p_i^* = -\frac{a_0 + u_i}{2a_1}
$$

And the optimal quantity will be: 

$$
q_i^* = a_0 + -a_1\frac{a_0 + u_i}{2a_1} + u_i = \frac{a_0 + u_i}{2}
$$

2) Calculate the OLS estimator

Now, we can insert it into the OLS estimator equation: 

$$
\hat a_1 = \frac{Cov(q_i^*, p_i^*)}{var(p_i^*)} = \frac{Cov(\frac{a_0 + u_i}{2}, -\frac{a_0 + u_i}{2a_1})}{var(-\frac{a_0 + u_i}{2a_1})}\\
= \frac{Cov(\frac{a_0}{2} + \frac{u_i}{2}, -\frac{a_0}{2a_1} - \frac{u_i}{2a_1})}{var(-\frac{a_0}{2a_1} - \frac{u_i}{2a_1})} \text{all are constants except $u_i$}\\
= \frac{Cov(\frac{u_i}{2}, -\frac{u_i}{2a_1})}{var(\frac{u_i}{2a_1})}\\
= \frac{-\frac{1}{4a_1}Cov(u_i, u_i)}{\frac{1}{4a_1^2}var(u_i)}\\
= - a_1
$$

This indicates that, as long as the monopolist has absolute price shock awareness, which is indicated that he is aware of the error term $u_i$, she can retrieve the optimal OLS estimator by simply multiplying the demand equation with -1 to obtain the correct price. However, we must also acknowledge that this scenario is already biased. 

### b) OLS Estimator when marginal costs are non-zero

In case of non-zero MC, we have an additional RV within the equation system. In this case, the profit maximization problem of the monopolist looks like: 

$$
max \underbrace{\pi}_{p_i} = a_0 + 2a_1p_i - c_ia_1 + u_i = 0
$$

1) Calculate the optimal price and quantity

Again, we retrieve: 

$$
p_i^* = \frac{c_ia_1 - a_0 - u_i}{2a_1}
$$

And the optimal quantity will be: 

$$
q_i^* = a_0 + a_1\frac{c_ia_1 - a_0 - u_i}{2a_1} + u_i = \frac{c_ia_1 + a_0 + u_i}{2}
$$

2) Calculate the OLS estimator

$$
\hat a_1 = \frac{Cov(q_i^*, p_i^*)}{var(p_i^*)} = \frac{Cov(\frac{c_ia_1 + a_0 + u_i}{2}, \frac{c_ia_1 - a_0 - u_i}{2a_1})}{var(\frac{c_i - a_0 - u_i}{2a_1})}\\
= \frac{Cov(\frac{c_ia_1}{2} + \frac{a_0}{2} + \frac{u_i}{2}, \frac{c_ia_1}{2a_1} -\frac{a_0}{2a_1}  -\frac{u_i}{2a_1})}{var(\frac{c_ia_1}{2a_1} -\frac{a_0}{2a_1} - \frac{u_i}{2a_1})} |\text{all are constants except $u_i$ and $c_i$}\\
= \frac{Cov(\frac{c_ia_1}{2} + \frac{u_i}{2}, \frac{c_ia_1}{2a_1} -\frac{u_i}{2a_1})}{var(\frac{c_ia_1}{2a_1} - \frac{u_i}{2a_1})}\\
=\frac{Cov(\frac{c_ia_1}{2a_1},\frac{c_ia_1}{2}) - Cov(\frac{u_i}{2}, \frac{u_i}{2a_1}) + Cov(\frac{u_i}{2}, \frac{c_ia_1}{2a_1}) - Cov(\frac{u_i}{2a_1}, \frac{c_ia_1}{2})}{var(\frac{c_ia_1}{2a_1}) - Var(\frac{u_i}{2a_1})} |\text{Cov($u_i$, $c_i$) = 0}\\
= \frac{\frac{a_1^2}{4a_1}Var(c_i) + \frac{1}{4a_1}Var(u_i)}{\frac{a_1^2}{4a_1^2}Var(c_i) + \frac{1}{4a_1^2}Var(u_i)}\\
= \frac{\frac{a_1}{4}Var(c_i) + \frac{1}{4a_1}Var(u_i)}{\frac{1}{4}Var(c_i) + \frac{1}{4a_1^2}Var(u_i)}\\
= a_1 - \frac{2a_1Var(u_i)}{a_1^2Var(c_i) + Var(u_i)}
$$

As we can see, also this term is biased as $\hat a_1 \neq a_1$

1. $h_{ii} = \frac{\hat y_i}{y_i}$

2. $\hat y_i = x_i\hat\beta_i$

3. $\hat\beta_i = (X'X)^{-1}x_i'y_i$

4. $\hat y_i = x_i(X'X)^{-1}x_i'y_i$

5. $h_{ii} = x_i(X'X)^{-1}x_i'$

Therefore, it solely depends on the x's! 

6. $\hat y_i = h_{i,1}y_1 + ... + h_{1,n}y_n = \sum^n_{i=1}h_{i,i}y_i$

$h_{ii}$ = Diagonal $i^{th}$ element of the Hat Matrix. The Hat Matrix summarizes all individual leverage factors. Each entry (Eigenvalue) of said matrix indicates the distance of the i^{th} observation from the center of x space. ¨¨

By multiplying the Eigenvector values with the respective observed y elements ($y_i$), we calculate each estimate of $y_i$ ($\hat y_i$). This matrix is supposedly an Eigenmatrix. 

In general, the leverage tells us for **each datapoint the influence the observed response has on its predicted value**. As it is a distance measures, it can serve as an indicator for extreme values and indicate outliers potentially. However, a large leverage does not automatically imply outlier position, just a larger difference from the mean of the x space. 

In general, the average leverage is: 

\begin{equation}
\bar h_{i,i} = \frac{\sum h_{i,i}}{N} \\ 
= \frac{rank(H)}{N}^* \\
= \frac{rank(X)}{N}^*\\
= \frac{trace(H)}{N}\\
= \frac{P}{N}
\end{equation}

*rank is the maximum number of linearily separable rows per matrix
*trace is the sum of diagonal elements, which, per definition, is the rank in an Eigenmatrix
*all Eigenvalues sum up to the number of covariates within a matrix 

Lastly, when the dimension of x is large relative to n, then we automatically increase the critical value for representing an observation as an extreme value based on its average leverage. In practice, we usually take a value of $2*\frac{P}{N}$ as indication for a possible outlier. However, if there are many points surrounding the mean but are in general imprecise, this bears larger potential to falsely exclude an indication.

```python
# Figure 3.13
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,5)) 
fig.suptitle('Figure 3.12')



# Plot 1
ax1.scatter(x=X, y=y, facecolors='None', edgecolors='k', alpha=.5) 
ax1.scatter(x=X[[20,41,7]], y=y[[20,41,7]], facecolors='None', edgecolors='r', alpha=1) 
sns.regplot(x=X, y=y, ax=ax1, order=1, ci=None, scatter=False, line_kws={'color':'r', 'lw':1})
sns.regplot(x=X_small, y=y_small, ax=ax1, order=1, ci=None, scatter=False, line_kws={'color':'b', 'lw':1})
ax1.set_xlabel('X'); ax1.set_ylabel('Y'); ax1.axis(xmax=4.5);
ax1.legend(['With obs. 20,41', 'Without obs. 20,41']);

# Highlight points
ax1.annotate("20", (1.1, 7), color='r')
ax1.annotate("7", (4, 5), color = "r")
ax1.annotate("41", (3.6, 12), color='r');



# Plot 2
ax2.scatter(x=h, y=e, facecolors='None', edgecolors='k', alpha=.5)
ax2.set_xlabel('Leverage'); ax2.set_ylabel('Residuals'); ax2.axis(xmax=0.25)
ax2.hlines(0,xmin=ax2.xaxis.get_data_interval()[0],
           xmax=ax2.xaxis.get_data_interval()[1], linestyles='dotted',color='k')

# Highlight outlier
ax2.scatter(x=h[[20,41,7]], y=e[[20,41,7]], facecolors='None', edgecolors='r', alpha=1);

# Highlight points
ax2.annotate("20", (0, 3.6), color='r')
ax2.annotate("41", (0.14, 0.2), color='r');
ax2.annotate("7", (0.14, 5), color = "r");
```

As we have seen, being an outliers or having high leverage alone might be not enough to conclude that an observation is influential.

What really matters is a combination of both: observations with high leverage and high residuals, i.e. observations that are not only different in terms of observables (high leverage) but are also different in terms of their relationship between observables and dependent variable (high residual).

### 3.3.3 Potential Problems

#### Non-linearities

How can we assess if there are non-linearities and of which kind? We can look at the residuals.

If the residuals show some kind of pattern, probably we could have fit the line better. Moreover, we can use the pattern itself to understand how.

```python
# Linear fit

X = cars.horsepower.values.reshape(-1,1)
y = cars.mpg

regr = lm.fit(X, y)

cars['pred1'] = regr.predict(X)
cars['resid1'] = cars.mpg - cars.pred1

# Quadratic fit
X2 = cars[['horsepower', 'hp2']]
regr.fit(X2, y)

cars['pred2'] = regr.predict(X2)
cars['resid2'] = cars.mpg - cars.pred2
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,5))
fig.suptitle('Figure 3.9')

# Left plot
sns.regplot(x=cars.pred1, y=cars.resid1, lowess=True, 
            ax=ax1, line_kws={'color':'r', 'lw':1},
            scatter_kws={'facecolors':'None', 'edgecolors':'k', 'alpha':0.5})
ax1.hlines(0,xmin=ax1.xaxis.get_data_interval()[0],
           xmax=ax1.xaxis.get_data_interval()[1], linestyles='dotted')
ax1.set_title('Residual Plot for Linear Fit')

# Right plot
sns.regplot(x=cars.pred2, y=cars.resid2, lowess=True,
            line_kws={'color':'r', 'lw':1}, ax=ax2,
            scatter_kws={'facecolors':'None', 'edgecolors':'k', 'alpha':0.5})
ax2.hlines(0,xmin=ax2.xaxis.get_data_interval()[0],
           xmax=ax2.xaxis.get_data_interval()[1], linestyles='dotted')
ax2.set_title('Residual Plot for Quadratic Fit')

for ax in fig.axes:
    ax.set_xlabel('Fitted values')
    ax.set_ylabel('Residuals')
```
